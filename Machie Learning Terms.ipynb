# Evaluation Metrics

# Mean Absolute error

It calculates the average of the absolute differences between the actual and predicted values. It gives an idea of how much the predictions deviate from the actual values on average, without considering the direction of the error

# Mean Square Error

It calculates the average of the squared differences between actual and predicted values.

# Root Mean square Error

It is the square root of the MSE, bringing it back to the same unit as the original data. It provides an interpretation in terms of the original data scale.

# R-square

It measures the proportion of variance in the dependent variable that is predictable from the independent variables

# Evaluation metrics in classification

Precision, Recall, F1 score, Accuracy

Confusion Matrix

Precision: It measures the accuracy of positive predictions. It tells you how many of the items identified as positive are actually positive.

precision = TP / (TP + FP)

Recall:Recall is the proportion of true positive predictions (TP) out of all actual positives (TP + FN)

recall =TP/ (TP+FN)

F1-Score= 2*(p+r)/(p+r)
Harmonic mean of precision and recall

accuracy = correct predictions/total predictions
